```python
import matplotlib.pyplot as plt

# Graph types and #nodes
graph_types = [
    "ChainMM",
    "FFNN",
    "Llama-block",
    "Llama-layer",
    "8GPU transformer block",
    "8GPU full transformer"
]

num_nodes = [112, 192, 215, 261, 356, 417]
x_labels = [f"{n} - {g}" for n, g in zip(num_nodes, graph_types)]

# DOPPLER data
doppler_inference_cost = [
    1.176693964,
    2.874622226,
    2.456788826,
    3.285271525,
    4.633890104,
    5.527813816
]

doppler_gpu_memory_footprint = [
    0.01,
    0.01,
    0.01,
    0.01,
    0.01,
    0.01
]

# Placeto data
placeto_inference_cost = [
    2.240775275,
    5.629495716,
    5.711405754,
    7.157622743,
    9.649270129,
    12.93221591
]

placeto_gpu_memory = [
    20,
    70,
    60,
    100,
    160,
    240
]

# Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Primary Y-axis: Inference Cost
color1 = 'tab:blue'
ax1.set_xlabel('Graph Type and # Nodes')
ax1.set_ylabel('Inference Cost Per Episode (Sec)', color=color1)
ax1.plot(x_labels, doppler_inference_cost, marker='o', label='DOPPLER - Time per episode', color=color1)
ax1.plot(x_labels, placeto_inference_cost, marker='s', label='Placeto - Time per episode', color='tab:cyan')
ax1.tick_params(axis='y', labelcolor=color1)
plt.xticks(rotation=30)

# Secondary Y-axis: GPU Memory
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('GPU Memory (MB)', color=color2)
ax2.plot(x_labels, doppler_gpu_memory_footprint, marker='o', linestyle='--', label='DOPPLER - GPU Mem Footprint', color=color2)
ax2.plot(x_labels, placeto_gpu_memory, marker='s', linestyle='--', label='Placeto - GPU Mem Footprint', color='tab:pink')
ax2.tick_params(axis='y', labelcolor=color2)

# Combine legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

# Title and layout
plt.title('Inference Cost and GPU Memory Footprint vs. Graph Type and Node Count')
fig.tight_layout()
plt.show()
```
